```python
# 전화번호 크롤링 -- (2) using select 

import requests
from bs4 import BeautifulSoup 

#  원하는 url에의 html 코드 받기 
response = requests.get("https://workey.codeit.kr/orangebottle/index")

# BeautifulSoup 타입으로 변환 

soup = BeautifulSoup(response.text,"html.parser")

phone_numbers = soup.select("span.phoneNum")

phone = []

for num in phone_numbers:
    phone.append(num.string)
phone
```

```python
# 상품 있는 페이지만 가져오기 

import time 
import requests
from bs4 import BeautifulSoup 

page = 1

headers = {
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36"
}


pages = []
while True:  
    url = "http://www.ssg.com/search.ssg?target=all&query=n&page="+str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    if len(soup.select(".csrch_tip")) == 0:
        pages.append(soup)
        print(str(page)+"번째 페이지 출력")
        page = page +1
        time.sleep(3)
    else:
        break

print(pages)

```

```python
# TV 시청률 크롤링 최종 프로젝트 
import pandas as pd
import time 
import requests 
from bs4 import BeautifulSoup

years = range(2010,2012)
months = range(1,3)
weeks = range(1,3)


pages = []
for year in years:
    for month in months:
        for week in weeks:           
                url = f"https://workey.codeit.kr/ratings/index?year={year}&month={month}&weekIndex={week}"
                response = requests.get(url)
                soup = BeautifulSoup(response.text, "html.parser")  
                
                if len(soup.select("tr.row")) > 1:
                    pages.append(soup)
                    
                records = []

                for page in pages:
                    순위 = page.select("td.rank")
                    채널 = page.select("td.channel")
                    프로그램 = page.select("td.program")
                    시청률 = page.select("td.percent")


                    for i in range(10):
                        record = []
                        record.append(순위[i].text)
                        record.append(채널[i].text)
                        record.append(프로그램[i].text)
                        record.append(시청률[i].text)
                        records.append(record)
                        
df = pd.DataFrame(data=records, columns=['rank', 'channel', 'program', 'rating'])
df 
                   
       
```
